## Now handle the missing values

```python
updated_df = df.dropna(axis=1)
```

```python
updated_df
```

### but deleting rows more preferred

```python
updated_df1 = df.dropna(axis = 0)
updated_df1
```

```python
updated = df
updated["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
updated
```

```python
updated.info()
```

```python
updated = df
updated["Gender"] = df["Gender"].fillna(df["Gender"].mode())
```

```python
updated
```

```python
updated.info()
```

## Now backward and forward fill

```python
df1 = df
```

```python
df1["Age"] = df1["Age"].bfill()
```

```python
df1
```

```python
df2 = df
```

```python
df2["Gender"] = df2["Gender"].ffill()
```
