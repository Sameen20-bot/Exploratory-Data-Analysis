```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import preprocessing
```

```python
df = pd.read_csv("Churn_Modelling.csv")
```

```python
df.head()
```

```python
df.info()
```

```python
df["Gender"].mode()
```

```python
df["Gender"] = df["Gender"].fillna("Male")
```

```python
df["Age"] = df["Age"].fillna(df["Age"].mean())
```

```python
df.info()
```

```python
df.drop(columns=["RowNumber","CustomerId","Surname"],inplace=True,axis=1)
```

```python
df.head()
```

# Label Encoding

```python
le = preprocessing.LabelEncoder()
df["Gender_Label"] = le.fit_transform(df.Gender.values)
```

```python
df.head()
```

# One Hot Encoding

```python
one_hot = pd.get_dummies(df["Geography"],dtype=int)
one_hot
```
